## Import modules

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')

%matplotlib inline
```

## Data reading

```python
df = pd.read_csv('Simple Linear Regression Practice.csv')
df.head()
```

## Size of data

```python
df.shape
```

## Checking missing values

```python
sns.heatmap(df.isna())
```

## Checking data types

```python
df.dtypes
```

## Creating variables for regression model

```python
X = df[['X']]
y = df['Y']
```

```python
X
```

```python
y
```

## Data plot

```python
plt.scatter(df['X'], df['Y'], color='g', alpha=0.2)
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Initial data")
```

## Splitting data to train and test set

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size = 0.3, random_state = 0)
xtrain
```

## Linear regression modelling on train data

```python
md = LinearRegression()
md.fit(xtrain, ytrain)
```

## Test data prediction

```python
md.predict(xtest)
```

## Checking predicted values

```python
plt.scatter(df['X'], df['Y'], color='g', alpha=0.2)
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Linear Regression")
plt.plot(df.X, md.predict(df[['X']]))
```

## Slope, intercept and R-squared

```python
slope = md.coef_[0]
print(f"slope: {slope}")
```

```python
intercept = md.intercept_
print(f"intercept: {intercept}")
```

```python
r_sq = md.score(X,y)
print(f"coefficient of determination: {r_sq}")
```

## Making prediction table

```python
prediction_table = df.copy()
```

```python
prediction_table['y_pred'] = md.predict(X)
prediction_table['Y - y_pred'] = prediction_table['Y'] - prediction_table['y_pred']
```
